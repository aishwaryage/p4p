# Exercise #1: Defining the derivative

```python
def fin_deriv(f, x, a):
	"""
	this defines the rate of change of the function f as a function of x

	the final point is at xf = x + a
	the initial point is at xi = x - a
	the difference between these two is xf - xi = 2*a
	"""

	return ((f(x+a) - f(x-a))/(2.0*a))
```

You can play around with this to understand the impact of changing "a" -- finer "resolution" gets more accurate results.

```python
x = np.arange(0.1, 5, 0.01)

a = 0.2 # try values between 0.1 and 2

# define sine function
def sin(x):
  f = np.sin(x)
  return f

# take the derivative
dsin = fin_deriv(sin, x, a)

plt.plot(x, np.sin(x), label="sin")
plt.plot(x, dsin, color='orange', label = "numerical derivative")
plt.plot(x, np.cos(x), '--', color='orange', label="actual derivative")
plt.xlabel('x')
plt.legend()
plt.show()
```

# Exercise #2: Test convergence

Now let's make some plots as a function of the resolution "a" and see what our error function looks like.

```python
# define our error function
def error_fun(Xvec,Yvec):

  """ This function takes a dependent variable (in our case, the derivative) and independent variable (in our case, a) and returns the error function. """

  num = Yvec[0:-1]-Yvec[1:] #gives an array of the difference of y at various points evaluated at x
  dnum = Xvec[0:-1]-Xvec[1:]#gives an array of the constant intervals of the x values sampled
  Ytmp = 100 * abs(num/dnum)

  Xtmp = Xvec[0:-1]

  return Xtmp, Ytmp

def master_plot_vs_a():

  """ This will format our plots for us, so we don't have to type this out every time. """

  plt.figure(figsize=(9.0,5))
  plt.subplot(211)
  plt.xticks(fontsize=10)
  plt.yticks(fontsize=10)

  plt.ylabel(ylabel1,size=15)

  plt.axhline(y=actual_deriv,color='k',linewidth=1, label="Actual Derivative")# the value of the derivative at x=1
  plt.plot(a, numerical_deriv, color='r', label="Numerical Derivative")#the value of the derivative evaluated at a particular 'a'
  plt.legend()

  plt.axvline(x=0,color='k',linewidth=1)

  plt.subplot(212)
  plt.xticks(fontsize=10)
  plt.yticks(fontsize=10)

  plt.axhline(y=0,color='k',linewidth=1)
  plt.axvline(x=0,color='k',linewidth=1)

  plt.ylabel(ylabel2 , size=15)
  plt.plot(xtmp, error_val, color='g', label="Error")
  plt.legend()


  plt.xlabel(xlabel0,size=25, position=(1,1.2))
  plt.show()


x = 1
a=np.arange(pow(10,-5),.1,pow(10,-3))


"""
******************************
******************************
    exponential
******************************
******************************
"""
ylabel1 = 'Derivative'
ylabel2 = 'Convergence Error'
xlabel0 = 'a'
xtmp = a[:-1]
actual_deriv = exp(x)


# your code here
# define exp function
def exp(x):
  return np.exp(x)
# take the numerical derivative of the exponential function
numerical_deriv = fin_deriv(exp, x, a)
# calculate the error function
x_err, error_val = error_fun(a, numerical_deriv)#array of error values for the derivative of the
#function as the value of a changes showing convergence for smaller values of a

# plot the numerical derivative and error
master_plot_vs_a()
```
