```python
def draw_graph(x, y, x_min, x_max, sol):
  plt.plot(x, y, markersize=.8)
  plt.xlabel("X-Position")
  plt.ylabel("Y-Position")
  plt.xlim(x_min, 1.5*x_max)
  plt.title('Ball Trajectory')
```

```python
def draw_path(v, sol):
  g = 9.8
  x = 2
  y = 4
  x_pan = np.arange(0, x, 0.01)
  theta = 0
  sol1 = np.arctan((x + sqrt(x**2 - (2*g*x**2/v**2) * (g*x**2/v**2 + y)))/(g*x**2/v**2))
  sol2 = np.arctan((x - sqrt(x**2 - (2*g*x**2/v**2) * (g*x**2/v**2 + y)))/(g*x**2/v**2))
  if sol == 1:
    theta = sol1
  else:
    theta = sol2

  y = math.tan(theta)*x_pan - 0.5*g*(x_pan/(v*math.cos(theta)))**2

  draw_graph(x_pan, y, min(x_pan), max(x_pan), sol)
```

```python
sol_list = [1, 2]
vel = 12
for s in sol_list:
  draw_path(vel, s)

plt.legend(['Solution 1', 'Solution 2'])

```
